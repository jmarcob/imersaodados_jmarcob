```python
import pandas as pd
url = "https://gist.githubusercontent.com/tgcsantos/3bdb29eba6ce391e90df2b72205ba891/raw/22fa920e80c9fa209a9fccc8b52d74cc95d1599b/dados_imoveis.csv"
dados = pd.read_csv(url)
dados
```

```python
dados_metragem = dados[["Bairro", "Metragem"]]
dados_metragem
```

```python
#Descobrindo a média das metragens por bairro
dados_metragem.groupby("Bairro").mean()
```
